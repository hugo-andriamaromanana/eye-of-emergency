```python
from pathlib import Path
from pandas import DataFrame, notna, read_csv

from scripts.ptdf import ptdf
from scripts.tweet import Tweet, create_tweet


def expand_tweet_df(tweet_data: DataFrame) -> DataFrame:
    """Returns a df with extra labels"""
    df_none = tweet_data.where(notna(tweet_data), None)
    obj_tweets: list[Tweet] = []
    for _, row in df_none.iterrows():
        tweet = create_tweet(row)  # type: ignore
        obj_tweets.append(tweet)
    return ptdf(obj_tweets)


_COLS_TO_DROP = [
    "keyword",
    "location",
    "txt",
    "txt_len",
    "nb_of_words",
    "hashtags",
    "urls",
    "usernames",
    "emails",
    "has_hashtag",
    "has_username",
    "has_email",
]


def clean_tweet_df(tweet_path: Path | str) -> None:
    TWEET_DATA = read_csv(tweet_path)
    TWEET_EXTRA = expand_tweet_df(TWEET_DATA)
    cleaned = TWEET_EXTRA.drop(columns=_COLS_TO_DROP)
    cleaned.to_csv("../data/train/test_tweets.csv", index=False)


TRAINING_DATASET_PATH = Path("../data/raw/train_tweets.csv")
TEST_DATASET_PATH = Path("../data/raw/test_tweets.csv")
clean_tweet_df(TRAINING_DATASET_PATH)
```
